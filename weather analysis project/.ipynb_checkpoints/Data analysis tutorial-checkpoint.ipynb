### What is the Mean value of each column against each 'Weather Condition ?

```python
groups=df.groupby('Weather_Condition')
for group_name , group_df in groups:
    print(group_name)
    print(group_df.describe())
    print("=========================")
```

```python
groups=df.groupby('Weather_Condition')
for group_name , group_df in groups:
    print("Condition ===",group_name)
    print(group_df.mean())
    print("=========================")
```

```python
# =================================== 2nd Method=======================Accurate(Prefered)
df.groupby('Weather_Condition').mean()
```
