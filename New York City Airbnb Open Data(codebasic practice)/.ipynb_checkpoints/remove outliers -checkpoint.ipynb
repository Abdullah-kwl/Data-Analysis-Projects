## we will see how to remove outliers from our data


```python
import pandas as pd
df=pd.read_csv("AB_NYC_2019.csv")
df
```

```python
df2=df.price
df2.head(50)
```

```python
df2.describe()
```

```python
df2.median()
```

```python
df2.mean()
```

```python
df2.quantile(1)
```

```python
df2.quantile(0.50)
```

```python
df2.quantile(0.1)
```

```python
df2.quantile(0.01)
```

```python
df2.quantile(0.997)
# 99.7%
```

```python
min_thresold , max_thresold=df2.quantile([0.01,0.997])
min_thresold , max_thresold
```

```python
df_filtered=df[(df.price>min_thresold) & (df.price<max_thresold)]
df_filtered
```
