# What is the mean stop_duration ?

```python
df.head(10)
```

```python
df.dtypes
```

stop_duration is object type so we cannot directaly find the mean of string we will use other method

# df.stop_duration.value_counts()

```python
df.stop_duration=df.stop_duration.map({'0-15 Min':6.5 , '16-30 Min':23 , '30+ Min':45})
df.stop_duration
```

```python
df.stop_duration.mean()
```
