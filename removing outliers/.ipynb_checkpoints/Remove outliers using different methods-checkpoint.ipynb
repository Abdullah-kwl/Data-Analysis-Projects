Remove outliers using percentile technique first. Use [0.001, 0.999] for lower and upper bound percentiles
After removing outliers in step 1, you get a new dataframe.
On step(2) dataframe, use 4 standard deviation to remove outliers
Plot histogram for new dataframe that is generated after step (3). Also plot bell curve on same histogram
On step(2) dataframe, use zscore of 4 to remove outliers. This is quite similar to step (3) and you will get exact same result

# filtring outliers using percentile technique : given Use [0.001, 0.999]

```python
min_val , max_val = df.price_per_sqft.quantile([0.001, 0.999])

print(min_val)
print(max_val)
```

```python
# we will remove the following values
df[(df.price_per_sqft<min_val) | (df.price_per_sqft>max_val)]
```

```python
# we will remove the following 28 values
(df[(df.price_per_sqft<min_val) | (df.price_per_sqft>max_val)]).count()
```

```python
df2=df[(df.price_per_sqft>min_val) & (df.price_per_sqft<max_val)]
df2
```

before filtring 13200 rows × 7 columns  || after filtring 13172 rows × 7 columns     || (Total 28  outliers are remove)

# use 4 standard deviation to remove outliers Plot histogram for new dataframe

```python
plt.hist(df2.price_per_sqft , bins=20, rwidth=0.8)
plt.xlabel('Price per square foot')
plt.ylabel('count')
# plt.yscale('log')
plt.show()
```

```python
# now we will further remove uotliers from df2
df2
```

```python
df2.price_per_sqft.describe()
```

```python
mean=df2.price_per_sqft.mean()
mean
```

```python
std=df2.price_per_sqft.std()
std
```

```python
min_range=mean-(4*std)
max_range=mean+(4*std)
min_range ,  max_range
```

```python
df3=df2[(df2.price_per_sqft>min_range) & (df2.price_per_sqft<max_range)]
df3
```

last count 13172 rows × 7 columns now we have 13047 rows × 7 columns (removed=125) 

```python
plt.hist(df3.price_per_sqft , bins=20, rwidth=0.8)
plt.xlabel('Price per square foot')
plt.ylabel('count')
# plt.yscale('log')
plt.show()
```

#  use zscore of 4 to remove outliers.

```python
# we will take df2 and check again that it also re move 125 rows or not
df2
```

```python
df2.price_per_sqft.describe()
```

```python
# we will add the z-Score coloum in df

df2['zscore']= (df2.price_per_sqft-df2.price_per_sqft.mean())/df2.price_per_sqft.std()
df2
```

```python
df2.zscore.describe()
```

```python
df2[(df2.zscore<-4) | (df2.zscore>4)]
```

```python
df4=df2[(df2.zscore>-4) & (df2.zscore<4)]
df4
```

last count 13172 rows × 7 columns now we have 13047 rows × 7 columns (removed=125) 
Z-score filtring also remove 125 rows
